```python
nyanglerbass_maskedextra_df = pd.read_csv('/home/desbrium/Metis/FishingSpots/Data/nyanglerbass_maskedextra.csv')
```

```python
nyanglerbass_maskedextra_df = nyanglerbass_maskedextra_df[nyanglerbass_maskedextra_df['Post_Views'] != 'No Info']
```

```python
nyanglerbass_maskedextra_df[['Thread_Name','Thread_Views', 'Thread_Replies']].value_counts()
```

```python
nyanglerbass_maskedextra_df['Post_Views'] = nyanglerbass_maskedextra_df['Post_Views'].str.replace('K', '000').str.replace('M', '000000').astype(int)
nyanglerbass_maskedextra_df['Post_Replies'] = nyanglerbass_maskedextra_df['Post_Replies'].str.replace('K', '000').str.replace('M', '000000').astype(int)
```

```python
nyanglerbass_maskedextra_df[['Post_Year','Post_Views', 'Post_Replies']].groupby('Post_Year').agg({'Post_Views' : ['sum', 'mean'], 'Post_Replies' : ['sum', 'mean']}).round(0).sort_values([('Post_Views', 'mean'),
            ('Post_Replies',  'mean')], ascending = False)
```

```python
nyanglerbass_maskedextra_df[['Post_Month','Post_Views', 'Post_Replies']].groupby('Post_Month').agg({'Post_Views' : ['sum', 'mean'], 'Post_Replies' : ['sum', 'mean']}).round(0).sort_values([('Post_Views', 'mean'),
            ('Post_Replies',  'mean')], ascending = False)
```

```python
nyanglerbass_maskedextra_df[['Post_Season','Post_Views', 'Post_Replies']].groupby('Post_Season').agg({'Post_Views' : ['sum', 'mean'], 'Post_Replies' : ['sum', 'mean']}).round(0).sort_values([('Post_Views', 'mean'),
            ('Post_Replies',  'mean')], ascending = False)
```

```python
nyanglerbass_maskedextra_df[['Thread_Name','Post_Views', 'Post_Replies']].groupby('Thread_Name').agg({'Post_Views' : ['sum', 'mean'], 'Post_Replies' : ['sum', 'mean']}).round(0).sort_values([('Post_Views', 'mean'),
            ('Post_Replies',  'mean')], ascending = False)
```
